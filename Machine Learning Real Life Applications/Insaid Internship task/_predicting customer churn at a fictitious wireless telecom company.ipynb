#### Data Cleaning

```python
cleaned_data = dataset.drop(columns='customerID')
```

```python
cleaned_data.head()
```

```python
#Getting dummy variables from categorical variables
cleaned_data = pd.get_dummies(cleaned_data,columns=['PaymentMethod','InternetService','Contract'])
```

```python
cleaned_data.head()
```

```python
cleaned_data.info()
```

```python
cleaned_data = cleaned_data.drop(columns=['Contract_Two year','InternetService_No','PaymentMethod_Mailed check'])
```

```python
#Encoding categorical variable to numerical variable 
from sklearn.preprocessing import LabelEncoder
for column in cleaned_data.columns:
    if cleaned_data[column].dtype == np.number:
        continue
    cleaned_data[column] = LabelEncoder().fit_transform(cleaned_data[column])
```

```python
cleaned_data.info()
```

#### Correlation Plot with respect to Target Variable

```python
#Correlation plot
cleaned_data.drop(columns='Churn').corrwith(cleaned_data.Churn).plot.bar(figsize=(20,10),title='Correlation with Target Variable',fontsize=15,rot=90,grid=True)
```

#### Correlation Matrix

```python
#Correlation Matrix
sns.set(style='white')
corr = cleaned_data.drop(columns=['Churn']).corr()

mask=np.zeros_like(corr,dtype=np.bool)
mask[np.triu_indices_from(mask)]=True

f,ax=plt.subplots(figsize=(15,12))

cmap = sns.diverging_palette(220,10,as_cmap=True)

sns.heatmap(corr,mask=mask,cmap=cmap,vmax=0.3,center=0,square=True,cbar_kws={'shrink':0.5},annot=True)
```

##### From above two visualization, we get to know that gender,PhoneService,TotalCharges are having nearly zeroo correlation with target variable.
So we can drop them from the data.

```python
cleaned_data = cleaned_data.drop(columns=['gender','PhoneService','TotalCharges'])
```

```python
X = cleaned_data.drop(columns='Churn')
y = cleaned_data['Churn']
```

#### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

#### Splitting The Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

### Implement Machine Learning Models and Evaluating models

##### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(C=0.75,n_jobs=-1,solver='lbfgs',penalty='l2',random_state=0,verbose=5)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#K-Fold CV
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=10)
print('Accuracy: {:0.2f} % '.format(accuracies.mean()*100))
print('Standard Deviation: {:0.2f} % '.format(accuracies.std()*100))
acc_lr = round(accuracies.mean()*100,2)
std_lr = round(accuracies.std()*100,2)
```

##### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier(n_estimators=100,random_state=10)
classifier2.fit(X_train,y_train)
```

```python
y_pred2 = classifier2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm = confusion_matrix(y_test,y_pred2)
print(cm)
print(accuracy_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

```python
#K-Fold CV
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier2,X=X_train,y=y_train,cv=10)
print('Accuracy: {:0.2f} % '.format(accuracies.mean()*100))
print('Standard Deviation: {:0.2f} % '.format(accuracies.std()*100))
acc_rf = round(accuracies.mean()*100,2)
std_rf = round(accuracies.std()*100,2)
```
